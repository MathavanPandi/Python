```python
df = pd.read_csv('oil-company-doe.csv')  # import data into a dataframe
df
```

## **1. fillna method**

```python
df1 = df.fillna(0)  # it will change the null values into 0
df1
```

### **Sometimes having 0 is not a better guess**

```python
df1 = df.fillna({'x1':0, 'x2':0, 'x3':'No', 'x4':0, 'y':0})  # it will put the specified values of that columns
df1
```

### **Still didnt satisfied the result use "ffill" or "bfill" method**

```python
df1 = df.fillna(method="ffill")  # ffill = forward fill , it will fill the forward values
df1
```

```python
df1 = df.fillna(method="bfill")   # bfill = backward fill , it will fill the backward values
df1
```

## **2. Interpolate Method**

**For a better guessing**

```python
df1 = df.interpolate()
df1
```

## **3. Dropna Method**

```python
#Sometimes based on situation i need to drop the missing values from my dataset

df1 = df.dropna()
df1
```
